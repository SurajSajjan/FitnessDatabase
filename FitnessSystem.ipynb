```python
import sqlite3
from sqlite3 import Error
 

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
 
    return conn
```

```python
def select_option1(conn):
    """
    1. The branch 2 was established in the year.
    """
    cur = conn.cursor()
    print("DateYear")
    cur.execute("""SELECT DateYear
                FROM Branch
                WHERE ID = 2;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0])
```

```python
def select_option2(conn):
    """
    2. The number of members who have subscribed to half-yearly plans in branch 4.
    """
    cur = conn.cursor()
    print("Count")
    cur.execute("""SELECT count(*)
                FROM Billing 
                LEFT JOIN Subscription ON Billing.SubscriptionID = Subscription.ID
                LEFT JOIN Member ON Billing.MemberID = Member.ID
                WHERE DurationMonths = 6 and BranchID = 4;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0])
```

```python
def select_option3(conn):
    """
    3. The membership fee for a member with a yearly plan who has opted for the individual package.
    """
    cur = conn.cursor()
    print("MembershipFee")
    cur.execute("""SELECT MemberFee1 
                FROM Temp
                WHERE Name1 = "Individual"
                AND DurationMonths1 = 12
                AND MemberID1 = 4;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0])
```

```python
def select_option4(conn):
    """
    4. The number of trainers who have experience of 3 years and above.
    """
    cur = conn.cursor()
    print("Count")
    cur.execute("""SELECT COUNT(*)
                FROM Employee
                WHERE Specialization = "Trainer"
                AND Experience >= 3;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0])
```

```python
def select_option5(conn):
    """
    5. The count of students/regular citizens/ senior citizens.
    """
    cur = conn.cursor()
    print("Status,Count")
    cur.execute("""SELECT Status, COUNT(Status)
                FROM Member
                GROUP BY Status;""")
 
    rows = cur.fetchall()
    for row in rows:
        print(row[0],row[1])
```

```python
def select_option6(conn):
    """
    6. The total number of male and female members in each branch.
    """
    cur = conn.cursor()
    print("BranchID, Gender, Count")
    cur.execute("""SELECT BranchID, Gender, COUNT(Gender)
                FROM Member
                GROUP BY BranchID, Gender;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0],row[1],row[2])
```

```python
def select_option7(conn):
    """
    7. The gross income of the organization.
    """
    cur = conn.cursor()
    print("Gross Income")
    cur.execute("""SELECT SUM(MemberFee)
                FROM Billing
                WHERE StartDateYear = 2013;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0])
```

```python
def select_option8(conn):
    """
    8. The contact details of each branch.
    """
    cur = conn.cursor()
    print("ID, ContactPhone, ContactEmail")
    cur.execute("""SELECT ID, ContactPhone, ContactEmail
                FROM Branch;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0],row[1],row[2])
```

```python
def select_option9(conn):
    """
    9. The maximum weight of a member among all the members and his calculated BMI.
    """
    cur = conn.cursor()
    print("Weight, BMI")
    cur.execute("""SELECT MAX(Weight), BMI 
                FROM Member;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0],row[1])
```

```python
def select_option10(conn):
    """
    10. The number of members who have booked a 5:00 pm slot for Zumba on 12/08/2013.
    """
    cur = conn.cursor()
    print("Count")
    cur.execute("""SELECT COUNT(*) 
                FROM Booking 
                WHERE Class = "Zumba" 
                AND DateMonth = 8 
                AND DateDay = 12 
                AND DateYear = 2013
                AND StartTimeHour = 17
                AND StartTimeMin = 00;""")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row[0])
```

```python
def select_option13(conn):
    """
    13. Deleting record from Employee Table.
    """
    count = 0
    ID = input("Enter the Employee ID of the employee who's details are to be deleted (Please ensure to enter the right ID) : ")

    cur = conn.cursor()
    cur.execute("""SELECT EXISTS(SELECT * FROM Employee WHERE ID=? LIMIT 1);""",(ID,))
    exist = cur.fetchall()
    for exi in exist:
        if exi[0] == 1:
            cur.execute("""DELETE FROM Employee WHERE ID = ?;""",(ID,))
            cur.execute("""DELETE FROM SQLITE_SEQUENCE WHERE NAME = "Employee";""")
            print("""\n This ID has been deleted. \n""")
            rows = cur.fetchall()
            for row in rows:
                print(row)
        else:
            print("""\n This is not a valid ID of the employee. \n""")
        
```
